```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df=pd.read_parquet('preprocessed_tweet_data.parquet')
```

```python
df.head()
```

```python
df[['user_followers_count', 'user_following_count', 'favorite_count', 'reply_count', 'retweet_count', 'quote_count', 'count_words', 'count_verbs']].describe()
```

```python
top_users = (
    df.groupby('user_screen_name', as_index=False)['user_followers_count']
    .max()
    .nlargest(10, 'user_followers_count')
)

top_users['user_followers_count_m'] = top_users['user_followers_count'] / 1_000_000

plt.figure(figsize=(12, 6))
ax = sns.barplot(
    x='user_screen_name',
    y='user_followers_count_m',
    data=top_users,
    palette='viridis'
)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}M',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10)

plt.title('Top 10 Users by Followers')
plt.ylabel('Number of Followers (Millions)')
plt.xlabel('User')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
top_users = (
    df.groupby('user_screen_name', as_index=False)['user_followers_count']
    .count()
    .nlargest(10, 'user_followers_count')
)

plt.figure(figsize=(12, 6))
ax = sns.barplot(
    x='user_screen_name',
    y='user_followers_count',
    data=top_users,
    palette='viridis'
)

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10)

plt.title('Top 10 Users by Number of Posts')
plt.ylabel('Number of Posts')
plt.xlabel('User')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
m = 0
for ws in df['words']:
    ld = len(ws)
    if ld > m:
        m = ld

words = []
positions = []
for ws, ps in zip(df['words'], df['positions']):
    word = list(ws) + [''] * (m - len(ws))
    position = list(ps) + [''] * (m - len(ps))

    words.append(word)
    positions.append(position)

words_array = np.array(words)
positions_array = np.array(positions)
```

```python
ws = [w for word in words for w in word if w != '']
text = ' '.join(ws)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Most Frequent Words')
plt.show()
```

```python
# 'ADJ', 'ADP', 'ADV', 'AUX', 'CCONJ', 'DET', 'INTJ', 'NOUN', 'NUM', 'PART', 'PRON', 'PROPN', 'PUNCT', 'SCONJ', 'SYM', 'VERB', 'X'

def frequency_graph(words_array, position):
    indices = np.where(positions_array == position)
    words_position = words_array[indices[0], indices[1]]

    words_freq = np.unique(words_position, return_counts=True)
    words_freq_word = words_freq[0][np.argsort(words_freq[1])][::-1]
    words_freq_count = words_freq[1][np.argsort(words_freq[1])][::-1]

    plt.figure(figsize=(15, 4))
    sns.barplot(x=words_freq_word[:15], y=words_freq_count[:15], palette='magma')

    plt.title(f'Top 15 Most Frequent {position}')
    plt.xlabel('Frequency')
    plt.ylabel(position)
    plt.grid(axis='x')
    plt.show()
```

```python
frequency_graph(words_array, 'NOUN')
```

```python
frequency_graph(words_array, 'ADJ')
```

```python
frequency_graph(words_array, 'VERB')
```
